```python
import random
num_list=[]
```

```python
def stock_time_stack():
    cnt = 0
    for _ in range(1000):
        num_list.append(random.randint(1,100))
    a = num_list
    ans = [0]*len(a)

    stack = []
    stack.append([987654321,-1])
    # print(stack[-1])
    for i in range(len(a)): # O(n)
        while stack[-1][0]<a[i]:
            top = stack.pop() # O(1)
            ans[top[1]] = i-top[1]
            cnt+=1
        stack.append([a[i],i]) # O(1)
    for i in range(1,len(stack)):
        ans[stack[i][1]] = -1
    return cnt
```

```python
def stock_time_n2():#[1,3,2,5,4] => [1,2,1,-1,-1] O(n^2)
    cnt = 0
    for _ in range(1000):
        num_list.append(random.randint(1,100))
    a = num_list
    ans = [-1]*len(a)
    
    for i in range(len(a)):
        for j in range(i+1,len(a)):
            cnt+=1
            if a[i]<a[j]:
                ans[i] = j-i
                break
    return cnt
```

```python
# n =1000
# 평균 while 연산 50959번:50*n
means = 0
meann = 0
for k in range(1000):
    means += stock_time_stack()
#     if k%25==0:
#         meann += stock_time_n2()
print("스택을 이용한 경우: "+means/50)
# print("완전탐색한 경우: "+meann/2)
```
